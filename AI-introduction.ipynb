# Midjourney

  - [Midjourney](https://www.midjourney.com/app/)

是個圖片創造聊天機器人，它可外掛在Discord聊天平台上。

# 設定

- 註冊[Discord](https://discord.com/)  
- 註冊[Midjourney](https://docs.midjourney.com/docs/quick-start#2-subscribe-to-a-midjourney-plan)
- 登入你的Discord帳號，邀請[Midjourney聊天機器人](https://docs.midjourney.com/docs/quick-start#3-join-the-midjourney-server-on-discord)  
- 在Midjourney創立的newbiew或general聊天室開始聊天作圖

# 聊天作圖

```
\imagine
```
